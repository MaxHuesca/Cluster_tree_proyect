## Parsed the tabular format 
When the program do a multi fasta our objective is to create a distance matrix o disimilarity matrix to used as a object for clustering in order to genereta a tree of distaces, note that this metos is not for generating a filogentic tree. 

```python
#Import the packages  
import pandas as pd 
import numpy as np 

#read the tsv
Data_frame_raw=pd.read_csv("../results/citrocromeC_seq_hmmr_clean_aling_clean.tsv", sep="\t",names=["query", "subject","score"])

axis_matrix=list(Data_frame_raw.drop_duplicates(subset="query")["query"]) 

#we make the matrix using pivot 
Matrix_df=Data_frame_raw.pivot(
    index="query",
    columns="subject",
    values="score"
) 
Matrix_df=Matrix_df.reindex(index=axis_matrix, columns=axis_matrix)
#we make a symetric matrix 
matrix_sym = Matrix_df.copy()

columns_mtrx=list(matrix_sym.columns)

for i in matrix_sym.index:
    for j in columns_mtrx:
        if pd.notna(Matrix_df.loc[i,j]):
            matrix_sym.loc[j,i] = Matrix_df.loc[i,j]
        elif pd.notna(Matrix_df.loc[j,i]):
            matrix_sym.loc[i,j] = Matrix_df.loc[j,i] 
    columns_mtrx.remove(i)
            

for col in matrix_sym.columns: 
    #we asign the diagonal of the matrix in 0 
    matrix_sym.loc[col,col]=0 

#obtain the max value from de matrix 
max_score=matrix_sym.max().max()

matrix_sym=matrix_sym/max_score
matrix_sym=matrix_sym.fillna(0)
matrix_sym
```

```python
Data_frame_raw = pd.read_csv(
    "../results/citrocromeC_seq_hmmr_clean_aling_clean.tsv",
    sep="\t",
    names=["query", "subject", "score"]
)

axis_matrix = sorted(
    set(Data_frame_raw["query"]).union(set(Data_frame_raw["subject"]))
)

Matrix_df = Data_frame_raw.pivot(
    index="query",
    columns="subject",
    values="score"
)

Matrix_df = Matrix_df.reindex(index=axis_matrix, columns=axis_matrix)

# Simetrización robusta
matrix_sym = np.maximum(Matrix_df, Matrix_df.T)

# Diagonal en 0
np.fill_diagonal(matrix_sym.values, 0)

# Normalización
max_score = np.nanmax(matrix_sym.values)
matrix_sym = matrix_sym / max_score


matrix_sym
```
